```python
from collections import Counter

s = 'python programming apssdc'

# Count the frequency of each character in the string
character_counts = Counter(s)

# Print the frequency of each character
for char, count in character_counts.items():
    print(f"{char} - {count}")
```

```python
s = 'python programming apssdc'
frequency = {}

for char in s:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

for char, count in frequency.items():
    print(f"{char} - {count}")
```

#### reduce:
* reduce(function,seq) is used to apply a particular functiona passed in its arguments to all of the list elements in the sequence.
* first we need import functools module

 working procedure:
 
The reduce(fun,seq) function is used to apply a particular function passed in its argument to all of the list elements mentioned in the sequence passed along.This function is defined in functools module.

Working :

At first step, first two elements of sequence are picked and the result is obtained. Next step is to apply the same function to the previously attained result and the number just succeeding the second element and the result is again stored. This process continues till no more elements are left in the container. The final returned result is returned and printed on console.

```python
import functools
```

```python
# To print the sum of all elements..
L1 = [1,2,3,4]
print(functools.reduce(lambda a,b:a+b,L1))

'''term1 = 1
term2 = 2
term3 = term1+term2 => 1+2 => 3
term1 = term3
term2 = term3+term1 => 3+3 => 6'''
```

#### 2. multi-level inheritance 
- In multi- level inheritance more than one parent and more than one child


```python
class Grandparent:
    def gdisplay(self):
        print('I am from gp class')

class Parent(Grandparent):
    def pdisplay(self):
        print('I am from p class')

class Child(Parent):
    def cdisplay(self):
        print('I am from c class')

ch = Child()
ch.gdisplay()  # Call the gdisplay method using parentheses
ch.pdisplay()
```

#### 3.Hierachical inheritance 
- one parent class and more than one child class

```python
class Parent:
    def pdisplay(self):
        print('I am from p class')

class Child1(Parent):
    def c1display(self):
        print('I am from ch1 class')

class Child2(Parent):
    def c2display(self):
        print('I am from ch2 class')

c2 = Child2()
c2.pdisplay()
```

#### 4. Multiple inheritance 
- two or more parent class one child class

```python
class Father:
    def fdisplay(self):
        print('I am from father class')

class Mother:
    def mdisplay(self):
        print('I am from mother class')

class Child(Father, Mother):
    def cdisplay(self):
        print('I am from child class')

# Create an instance of the Child class
child = Child()

# Call the methods from each class
child.fdisplay()  # Output: I am from father class
child.mdisplay()  # Output: I am from mother class
child.cdisplay()  # Output: I am from child class
```

#####  5.hybrid inheritance
- is a combination of multiple types of inheritance, such as hierarchical inheritance and multiple inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(self.name, "is eating.")


class Mammal(Animal):
    def __init__(self, name):
        super().__init__(name)

    def feed_milk(self):
        print(self.name, "is feeding milk.")


class Bird(Animal):
    def __init__(self, name):
        super().__init__(name)

    def fly(self):
        print(self.name, "is flying.")


class Bat(Mammal, Bird):
    def __init__(self, name):
        super().__init__(name)


bat = Bat("Bobby")
bat.eat()  # Output: Bobby is eating.
bat.feed_milk()  # Output: Bobby is feeding milk.
bat.fly()  # Output: Bobby is flying.
```

```python
class A:
    def display_A(self):
        print("I am from class A")

class B(A):
    def display_B(self):
        print("I am from class B")

class C(A):
    def display_C(self):
        print("I am from class C")

class D(B, C):
    def display_D(self):
        print("I am from class D")

# Create an instance of class D
d = D()

# Call methods from different classes
d.display_A()  # Output: I am from class A
d.display_B()  # Output: I am from class B
d.display_C()  # Output: I am from class C
d.display_D()  # Output: I am from class D
```

#### data imputation (used to deal with mising values)  

* it is the method for retatinig the majority of datasets and information by substituting missing data with different value 
* if data set have so many missing values we dont use that dataset 
* if only a few values are missing we can prefrom dataa imputation to repalce missing data with other values 
* there are many different methods for data imputation 
  - by using the mean values 
  - by using the median values
  - by using most frequent value(mode)
  - filling missing data with a constant 
  


```python
from sklearn. impute import SimpleImputer 
```

```python
# data frame in which abc are row and 123 are colomuns  and put some missing values 
import pandas as pd
import numpy as np
di={
    "a" : pd.Series([10,89,76,np.nan,90],index=[1,2,3,4,5]),
     "b" : pd.Series([222,23,24,22],index=[1,2,3,4]),
     "c" : pd.Series([10,89,35,35],index=[1,2,3,4])
}
df = pd.DataFrame(di)
df
```

* fit_transform() from sklearn,preprocessing class is used to preprocessing data for model training
* fit() method calculates various required parameters and
* transform() method applies the calculated parameters to standardize the data
* the fit_transform() is used to fit the data into a model and transform it into a form that is more suitable for model in a single step.This saves our time and effort for calling both fill() and transform() separately

```python
si=SimpleImputer(strategy="median")
si.fit_transform(df)
```

```python
si=SimpleImputer(strategy="most_frequent")
si.fit_transform(df)
```

```python
si=SimpleImputer(strategy="constant",fill_value=-1)
si.fit_transform(df)
```
